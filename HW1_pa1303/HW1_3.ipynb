```python
import os
import sys

# you can use the numpy library for random number generation: 
# http://docs.scipy.org/doc/numpy/reference/routines.random.html

import numpy as np
```

```python
np.random.seed(1)
```

```python
ReprRand = np.random.randn(2,100)
ReprRand.shape
```

```python

ReprRandAll = np.zeros((50, 2, 100))
for i in range(50):
    ReprRandAll[i] = 4 * np.random.randn(2, 100) + 5
```

```python
ReprRandAll.shape
```

```python

import pylab as pl
%pylab inline
# the last line of code makes the plot show up within the notebook. 
# Always include it so that your homework notebooks are rendered and we can see
# what you wanted your results to be. 
for i in range(len(ReprRandAll)):
    pl.plot(ReprRandAll[i][0], ReprRandAll[i][1], 'o')
pl.title('Gaussian Randomness')
```

```python
X = np.random.randint(1, 10)
Y = np.random.randint(1, 10)
for i in range(len(ReprRandAll)):
    pl.plot(ReprRandAll[i][0] + X, ReprRandAll[i][1] + Y, 'o')
pl.title('More Gaussian Randomenss')
print(X, Y)
```
