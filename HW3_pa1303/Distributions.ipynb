```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

nsamples = []
psamples = []
csamples = []
bsamples = []
esamples = []
minv = 10
maxv = 2000

for i in range(100):
    a = np.random.normal(40, 1, np.random.randint(minv, maxv))
    nsamples.append(a)
    b = np.random.poisson(40, np.random.randint(minv, maxv))
    psamples.append(b)
    c = np.random.chisquare(40, np.random.randint(minv, maxv))
    csamples.append(c)
    d= np.random.binomial(100, .4, np.random.randint(minv, maxv))
    bsamples.append(d)
    e = np.random.exponential(40, np.random.randint(minv, maxv))
    esamples.append(e)
    
```

```python
nmeans = [np.mean(x) for x in nsamples]
pmeans = [np.mean(x) for x in psamples]
cmeans = [np.mean(x) for x in csamples]
bmeans = [np.mean(x) for x in bsamples]
emeans = [np.mean(x) for x in esamples]
```

```python
for i in nsamples:
    plt.scatter(len(i),np.mean(i))
plt.xlabel('experiment size')
plt.ylabel('mean')
plt.title('Normal Distribution Means')
    
plt.figure(figsize=(6, 4))
plt.hist(nmeans)
plt.xlabel('sample mean', fontsize = 18)
plt.ylabel('N', fontsize = 18)
```

```python
for i in psamples:
    plt.scatter(len(i),np.mean(i))
    
plt.xlabel('experiment size')
plt.ylabel('mean')
plt.title('Poisson Distribution Means')
plt.figure(figsize=(6, 4))
plt.hist(pmeans)
plt.xlabel('sample mean', fontsize = 18)
plt.ylabel('N', fontsize = 18)
```

```python

for i in csamples:
    plt.scatter(len(i),np.mean(i))
    
plt.xlabel('experiment size')
plt.ylabel('mean')
plt.title('Chisquare Distribution Means')
plt.figure(figsize=(6, 4))
plt.hist(cmeans)
plt.xlabel('sample mean', fontsize = 18)
plt.ylabel('N', fontsize = 18)
```

```python
for i in bsamples:
    plt.scatter(len(i),np.mean(i))
    
plt.xlabel('experiment size')
plt.ylabel('mean')
plt.title('Binomial Distribution Means')
plt.figure(figsize=(6, 4))
plt.hist(bmeans)
plt.xlabel('sample mean', fontsize = 18)
plt.ylabel('N', fontsize = 18)
```

```python
for i in esamples:
    plt.scatter(len(i),np.mean(i))
    
plt.xlabel('experiment size')
plt.ylabel('mean')
plt.title('Exponential Distribution Means')
plt.figure(figsize=(6, 4))
plt.hist(emeans)
plt.xlabel('sample mean', fontsize = 18)
plt.ylabel('N', fontsize = 18)
```

The Law of Large Numbers states that as the size of experiemnts increase,
the ratio of outcomes becomes closer to the mean. This is evident through all 
the scatter plots of the means. As N increases the bands of the means narrow and become
closer to the mean.
